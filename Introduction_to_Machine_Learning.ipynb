```python
#import numpy and pandas libraries
import numpy as np
import pandas as pd
#import the dataset from downloaded files
bike_hire_data=pd.read_csv("C:\\Users\\Mary Mbaire\\Downloads\\bike_rental.csv")
bike_hire_data.head()
```

```python
#use pandas' datetimeindex library to change date format to successive days, add 'day' column to the table
bike_hire_data['day'] = pd.DatetimeIndex(bike_hire_data['dteday']).day
bike_hire_data
```

```python
#temperature data
temperature=bike_hire_data['temp']
temperature.head()
```

```python
#rental quantity data
rentals=bike_hire_data['rentals']
rentals.head()
```

```python
#import matplot library to enable plotting graphs
import matplotlib.pyplot as plt
%matplotlib inline
#define figure size
plt.figure(figsize=(8,6))
#plot rentals against temperature
plt.scatter(temperature,rentals,alpha=0.5,color='blue')
plt.xlabel('average temperature of the day')
plt.ylabel('total rentals of the day')
plt.title('number of rentals against average temperature of the day')
# Display the plot
plt.show()
```

```python
#put temperature data in a numpy array
features = np.array(temperature)
print(len(features))
print(features)
```

```python
#put rental data in a numpy array
labels=np.array(rentals)
print(len(labels))
labels
```

```python
# Combine features and labels into a dataset with features and labels
dataset_combined = np.column_stack((features, labels))
dataset_combined
```

```python
#import linear regression model used to train the model
from sklearn.linear_model import LinearRegression
```

```python
#reshape features to change to 2D array that can be manipulated by sckit learn
features = features.reshape(-1, 1)
# Fit an  linear regression model
model = LinearRegression()
fitted_model= model.fit(features, labels)
slope = model.coef_[0]
intercept = model.intercept_
# Plot the scatter plot of labels vs features and the OLS trendline
plt.figure(figsize=(8, 6))
plt.scatter(features, labels, label="Data points", color='blue', alpha=0.7)
plt.plot(features, intercept + slope * features, color='red', label="Regression Trendline")

# Customize plot labels and title
plt.xlabel('average temperature of the day')
plt.ylabel('total rentals of the day')
plt.title('number of rentals against average temperature of the day')
plt.legend()

# Show the plot
plt.show()
```

```python
# Convert the temperature(feature) to a 2D array
temperature = np.array([0.55]).reshape(-1, 1)
# Use the model to predict number of rentals expected for a temperature of 0.55
approximate_rentals = fitted_model.predict(temperature)
# Print the result
print('Prediction: {:.0f} rentals'.format(np.round(approximate_rentals[0])))
```

```python

import joblib
# Save the model as a pickle file
filename = './bike-share.pkl'
joblib.dump(fitted_model, filename)
```

```python
# Load the model from the file
loaded_model = joblib.load(filename)
# Use the model to predict tomorrow's rentals
temperature_new = np.array([0.41]).reshape(-1, 1)
result = loaded_model.predict(temperature_new)
print('Prediction: {:.0f} rentals'.format(np.round(result[0])))
```
